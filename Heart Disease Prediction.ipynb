# Heart Disease Prediction

### Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import mean_squared_error,accuracy_score,confusion_matrix
```

### Loading dataset

```python
df=pd.read_csv(r"HeartAttack.csv",na_values="?")
```

```python
df=df.drop(columns=["slope","ca","thal"],axis=1)
```

```python
df.head()
```

```python
df=df.dropna()
df.head()
```

```python
df.isnull().sum()
```

```python
df["cp"].value_counts()#return the frequency of distinct elements
```

### Dummies using pandas to convert categorical value to one-hot encoding

```python
df=pd.get_dummies(df,columns=["cp","restecg"])
df.head()
```

```python
df.columns
```

```python
df=df.rename(columns={"num       ":"target"})
df.head()
```

```python
numerical_cols=["age","trestbps","chol","thalach","oldpeak"]
cat_cols=list(set(df.columns)-set(numerical_cols)-{"target"})
```

```python
df["oldpeak"].value_counts()
```

```python
cat_cols
```

```python
numerical_cols
```

### Splitting data

```python
df_train,df_test=train_test_split(df,test_size=0.2,random_state=42)
len(df_train),len(df_test)
```

```python
scaler=StandardScaler()

def get_features_and_target_arrays(df,numerical_cols,cat_cols,scaler):
    x_numeric_scaled=scaler.fit_transform(df[numerical_cols])
    x_categorical=df[cat_cols].to_numpy()#categorical cols
    x=np.hstack((x_categorical,x_numeric_scaled))
    y=df["target"]

    return x,y
```

```python
x_train,y_train=get_features_and_target_arrays(df_train,cat_cols,numerical_cols,scaler)
```

### Logistic Regression

```python
clf=LogisticRegression()
clf.fit(x_train,y_train)
```

```python
x_test,y_test=get_features_and_target_arrays(df_test,cat_cols,numerical_cols,scaler)
```

```python
test_pred=clf.predict(x_test)
print("Mean squared error",mean_squared_error(y_test,test_pred))
```

```python
print("Accuracy",accuracy_score(y_test,test_pred))
```

```python
confusion_matrix(y_test,test_pred)
```

### Decision tree


```python
dc_clf=DecisionTreeClassifier()
dc_clf.fit(x_train,y_train)

dlf_pred=dc_clf.predict(x_test)
print("Mean squared error",mean_squared_error(y_test,dlf_pred))
print("Accuracy",accuracy_score(y_test,dlf_pred))
```

### Random forest classifier

```python
rc_clf=RandomForestClassifier()
rc_clf.fit(x_train,y_train)

rlf_pred=rc_clf.predict(x_test)
print("Mean squared error", mean_squared_error(y_test,rlf_pred))
print("Accuracy", accuracy_score(y_test,rlf_pred))
```

### SVM classifier


```python
svm_clf=SVC()
svm_clf.fit(x_train,y_train)

svm_pred=svm_clf.predict(x_test)
print("Mean squared error",mean_squared_error(y_test,svm_pred))
print("Accuracy",accuracy_score(y_test,svm_pred))
```
